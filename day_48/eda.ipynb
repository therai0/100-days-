```python
"""
 Here we will do EDA and Feature Engineering
"""
import pandas as pd 
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# reading the excel file
df = pd.read_excel('flight_price.xlsx')
df.head()
df.tail(1)

```

```python
df.info()
```

```python
df.describe()
# there is only a neumerical value it only show the information about the price
```

```python
# Feature engineering 
# working the Date_of_Journey
df['Day']= df['Date_of_Journey'].str.split('/').str[0] #day
df['Month'] = df['Date_of_Journey'].str.split('/').str[1] #Month
df['Year'] = df['Date_of_Journey'].str.split('/').str[2] #Year
# still the datatypes of this years month and day is object so we need to convert this all in to numeric datatypes
# i.e int 
```

```python
df['Day']=df['Day'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
df.info()

# 13  Day              10683 non-null  int64 
#  14  Month            10683 non-null  int64 
#  15  Year             10683 non-null  int64 
# now here we can see that all the column value is int 
```

```python
# Now we have extract the value for Date_of_journey column and seprate the value to different and convert the 
# each datatypes into int 
# so this column is no more needed 

df.drop('Date_of_Journey',axis=1,inplace=True)
df.head(1)
# so now we drop the column 
```

```python
# Feature engineering with Arrival_Time
# write the lamda function 
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0].astype(int) #converting also in int data types
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1].astype(int)
df['Arrival_Minute']
df.head(1)
```

```python
# now droping Arrival_Time column
df.drop('Arrival_Time',axis=1,inplace=True)
df.head(1)
```

```python
# Feature engineering With Depature_Time
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0].astype(int)
df['Dep_Minute'] = df['Dep_Time'].str.split(':').str[1].astype(int)
df.head(1)
```

```python
# removing the column Dep_Time
df.drop('Dep_Time',axis=1,inplace=True)
df.head(1)
```

```python
# Feature engineering with Duration
df['Duration'] = df['Duration'].apply(lambda a:a.split(' '))
```

```python
df['Duration_Hour']=  df['Duration'][0][0][0]
df['Duration_Minute']=  df['Duration'][0][1][0]
df.info()
```

```python
# converting into int datatype
df['Duration_Hour']= df['Duration_Hour'].astype(int)
df['Duration_Minute']= df['Duration_Minute'].astype(int)
df.info()
```

```python
df.head(1)
```

```python
# Featuring Engineering
# working with categorial varaible
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
# this row have the NaN value 
```

```python
# Now in this categorial varaible we need to label 
# Because ML model can't understand the character so we need to conver into numberical value
# we use to  Ordinal Encoding 
# Because number of stops increase price also increase
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stops':1,'2 stops':2,'3 stops':3,'4 stops':4}).fillna(1)
```

```python
df.head(2)
# we replace the categorial value with neumeric value
```

```python
# this neumeric value is float so convert it into integer
df['Total_Stops'] = df['Total_Stops'].astype(int)
```

```python
# we have both destination and source so we don't need route
df.drop('Route',axis=1,inplace=True)
```
