```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np 
```

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
# split the train and test data 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
# model trainig 
from sklearn.svm import SVC
svc = SVC(kernel='linear') # data is clearly seperated so using the linear kernal
```

```python
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
conf = confusion_matrix(y_test,y_pred)
class_report = classification_report(y_test,y_pred)
print(conf)
print(class_report)
# around 99% accuracy
```

## Model in overlap dataset 


```python
x,y = make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=1,n_redundant=0,random_state=20)
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
# here dataset is not seperated so use the differen kernal which will give the best output
```

```python
# split the train and test data 
# from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
# kernal linear
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
k_linear = SVC(kernel='linear')
k_linear.fit(x_train,y_train)
y_pred = k_linear.predict(x_test)
report = classification_report(y_test,y_pred)
conf_metrix = confusion_matrix(y_test,y_pred)
print(report)
print(conf_metrix)
```

```python
# rbf kernal
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
rbf = SVC(kernel='rbf')
rbf.fit(x_train,y_train)
y_pred = rbf.predict(x_test)
report = classification_report(y_test,y_pred)
conf_metrix = confusion_matrix(y_test,y_pred)
print(report)
print(conf_metrix)
```

```python
# poly kernal
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
poly_kr = SVC(kernel='poly')
poly_kr.fit(x_train,y_train)
y_pred = poly_kr.predict(x_test)
report = classification_report(y_test,y_pred)
conf_metrix = confusion_matrix(y_test,y_pred)
print(report)
print(conf_metrix)
```

```python
# sigmoid kernal
# Literal['linear', 'poly', 'rbf', 'sigmoid',
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
sig_moid = SVC(kernel='sigmoid')
sig_moid.fit(x_train,y_train)
y_pred = sig_moid.predict(x_test)
report = classification_report(y_test,y_pred)
conf_metrix = confusion_matrix(y_test,y_pred)
print(report)
print(conf_metrix)
```
