```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
```

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples=100,n_features=1,n_targets=1,noise=20)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(x,y)
```

```python
print(f"Cofficent:{regression.coef_}")
print(f"Intercept:{regression.intercept_}")
```

```python
# so here is the coffiecent is 27.35581966
# so here is the intercet is 0.559135500995541 
# this create the best fit line 
```

```python
# now we will try to get this intercept manually and undertstand about the gradient descent
# epochs means number of times loop need  will run to get intercet for best fit line
class GradientDescent:
    def __init__(self,learning_rate,epochs):
        self.m = 74.49935552
        self.lr = learning_rate
        self.b = -50 # initial intercetp is 100 so we need to reach to the 0.559135500995541
        self.epochs = epochs

    def find_best_intercept(self,x,y):
        for i in range(self.epochs):
            loss_slope = -2 * np.sum(y - self.m*x.ravel()-self.b)
            self.b = self.b - (loss_slope*self.lr)
        print(f"The value of intercet is {self.b}")

```

```python
gd = GradientDescent(0.001,200)
gd.find_best_intercept(x,y)
```

 here we can see that after the 200 times of updating the intercept manually we nearly get the exect value.
