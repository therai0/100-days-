## Car Price predication

```python
import seaborn as sns 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('cardekho_imputated.csv')
dataset.head()
```

```python
# Data cleaning
dataset.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
# check null value 
dataset.isnull().sum()
```

```python
# dropping the unecessary column form the dataset
dataset.drop(['car_name','brand'],axis=1,inplace=True)
```

```python
dataset['model'].unique()
```

```python
x= dataset.drop(['selling_price'],axis=1)
y= dataset['selling_price']
```

```python
# split the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Feature Encoding and scaling

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer

lb_encoder = LabelEncoder()
x['model'] = lb_encoder.fit_transform(dataset['model'])
```

```python
dataset.head()
```

```python
cat_features = ['seller_type', 'fuel_type', 'transmission_type']
num_features = x.select_dtypes(exclude='object').columns
```

```python
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(drop='first')
```

```python
preprocessor = ColumnTransformer([
    ("OneHotEncoder",cat_transformer,cat_features),
    ("StandardScaler",num_transformer,num_features)
],remainder='passthrough')
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
pd.DataFrame(x_train)
```

## Model traning and model selection

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
def evaluate_model(actual_value,predicated_value):
    score = r2_score(actual_value,predicated_value)
    MAE = mean_absolute_error(actual_value,predicated_value)
    MSE = mean_squared_error(actual_value,predicated_value)
    RMSE = root_mean_squared_error(actual_value,predicated_value)
    return score,MAE,MSE,RMSE
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'KNeighborsRegressor':KNeighborsRegressor(),
    'RandomForestRegressor':RandomForestRegressor(),
    'DecisionTreeRegressor':DecisionTreeRegressor()
}
```

```python
for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(x_train,y_train)

    # make prediction for train and test data
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)

    # evaluating models
    tr_score,tr_MAE,tr_MSE,tr_RMSE = evaluate_model(y_train,y_train_pred)
    tst_score,tst_MAE,tst_MSE,tst_RMSE = evaluate_model(y_test,y_test_pred)

    print("\n")
    # list of models
    print("=="*30)
    print(list(models.keys())[i])

    print("Performance of model for Training dataset")
    print(f"R2 score:{tr_score}")
    print(f"Mean Absolute Error:{tr_MAE}")
    print(f"Mean Squared Error:{tr_MSE}")
    print(f"Root Mean Squared Error:{tr_RMSE}")

    print('--'*30)

    print("Performance of model for Training dataset")
    print(f"R2 score:{tst_score}")
    print(f"Mean Absolute Error:{tst_MAE}")
    print(f"Mean Squared Error:{tst_MSE}")
    print(f"Root Mean Squared Error:{tst_RMSE}")
```
