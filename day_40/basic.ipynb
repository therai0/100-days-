```python
import seaborn as sns


# loading the build-in data from seaborn lib
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
# check there is missing value of not 
df.isnull()
# if there is null value it show the true value else false
```

```python
# df.dropna()  #delete the row
```

```python
""" Imputation Techniques for Missing Values """
# Mean Value Imputation
# now we will replace the missing value with the mean value of particular column

# let's replace the missing value of age column by mean of age 
df['age_mean'] = df['age'].fillna(df['age'].mean())
```

```python
df[['age','age_mean']] #let's see null value is replced by mean or not
# NaN value is replace by mean
# This technique works well when the data is normally distributed.
```

```python
""" Median Value Imputation
If the data is skewed or contains outliers, median value imputation is preferred."""

# let's replace the NaN value with median
df['age_median'] = df['age'].fillna(df['age'].median())
```

```python
df[['age_median','age_mean','age']]
# we can see that NaN in age_median column is replace by median value
```

```python
"""
Mode Imputation for Categorical Variables
Mode imputation is used for categorical variables. 
"""
# in the embarket column there few value which are missing 
# it falls under the categorical variables
# there are differnt tyeps category in embarked column
# in the missing value of embarked we will fill with mode value
# let's find the mode value

mode_value = df['embarked'].mode()[0] #this is the mode value
```

```python
# now replace the NaN value with mode value
df['embarked_mode'] = df['embarked'].fillna(mode_value)
df[['embarked','embarked_mode']]
```

```python
# now check the if there is any null value
df['embarked_mode'].isnull().sum()
# so there is no null value
```
