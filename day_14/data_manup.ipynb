```python
# any cell have Null value it will return True
df.isnull()
# 99th row have one missing value it show True
```

```python
# it will return the boolean value for rows 
# if any rows that have missing value then it will return the True
df.isnull().any(axis=1)
# 99th row have missing value so it return the True in this row
# axis=1 means rows wise default axis=0 which means column wise
```

```python
# it return the total number of missing value in each column
df.isnull().sum()
# website column have one missing value so it show the number 1
```

```python
# there is a single mising value in website column 
# so we will fill that missing value with https://youtube.com

# data cleaning
df = df.fillna("https://youtube.com")
df.tail(1)
# now the NaN value of website in 99th row has been updated
```

```python
# chaning or rename the coumn name
df = df.rename(columns={'Website':'website_link'})
df.head(1)
# now the Website column name is updated to website_link
```

```python
# we can apply the function to the particular column 

# apply the function which will convert the last and first name to uppercase
df["Upper_firstname"] = df["First Name"].apply(lambda x : x.upper()) # we using the lamda function

# now for lastnmae
df["Upper_lastname"] = df["Last Name"].apply(lambda x : x.upper()) # we using the lamda function
df.head(1)
# at the end of this there is two new column name Upper_firstname and Upper_lastname and value in upper case
```

```python
# Data aggregrating and grouping

# Data aggregation is the process of combining multiple pieces of data to produce a summary or a single result — like sums, averages, counts, etc. It helps you get insights from large datasets by grouping and summarizing information.

# identify the number of cities in particular country
citys_coutnry = df.groupby('Country')["City"].count()
citys_coutnry.tail(5)

# here Zimbabwe have two cities
#  others have only one
```

```python
# counting the number of user per country and city

users_number = df.groupby(['Country','City'])["First Name"].count()
users_number.head(10)
```

```python
# reading the new file product.csv
product_df = pd.read_csv('product.csv')

# performing the different aggregration method 

# finding the means price  according category
mean_price_stationary = product_df.groupby(['Category'])['Price'].mean()
mean_price_stationary
```

```python
product_df.head(1)
# counting the number of items based  category
number_items = product_df['Category'].value_counts()
number_items.head(10)
```

```python

# using multiple aggregating function

multi_agg = product_df.groupby('Category')['Price'].agg(['mean','sum','count'])
multi_agg.head(10)
```

```python
# merging and joining dataframes

# dataframe one
df_1 = pd.DataFrame(
    [{'student':'Bhaskar','class':7,'grader':'A'},
    {'student':'Bishal','class':7,'grader':'A+'},]
)

# dataframe two
df_2 = pd.DataFrame(
    [{'student':'Jemish','class':7,'grader':'B+'},
     {'student':'Bishal','class':7,'grader':'A+'},
    {'student':'Aayush','class':7,'grader':'A'},]
)
df_1
df_2
```

```python
# inner joining the two dataframes

pd.merge(df_1,df_2,how='inner')
# Returns only the rows that have matching values in both tables.
```

```python
# outer join
pd.merge(df_1,df_2,how='outer')
# Outer joins include unmatched rows from one or both tables in addition to the matched rows.
```

```python
# left join
pd.merge(df_1,df_2,how='left')
# Returns all rows from the "left" table (the first table specified in the FROM clause).
# Returns only the matching rows from the "right" table.
```

```python
# right join
pd.merge(df_1,df_2,how='right')
# Returns only the matching rows from the "left" table
```
