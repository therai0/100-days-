```python
import pandas as pd

wine = pd.read_csv('winequality-white.csv',sep=';')

wine.tail()
wine.info()
```

```python
# to get all the information
wine.info()
```

```python
# To get descriptive statistics such as mean, standard deviation, min, max, and percentiles for each feature, we use:
wine.describe()
```

```python
# Dataset Shape and Columns
print(wine.shape) #shape
print(wine.columns)
```

```python
# To see the unique quality scores present in the dataset:
wine['chlorides'].unique()
# this all value are the unique in column
```

```python
# Checking for Missing Values
wine.isnull().sum()
# all are 0 means there is no any missing value 
```

```python
# Checking and Removing Duplicate Records

wine[wine.duplicated()]
# 937 rows × 12 columns this much data is duplicated
```

```python
# removing the all the duplicated from ta table 
wine.drop_duplicates(inplace=True)
```

```python
wine.shape
# initially (4898, 12)
# currently (3961, 12)
```

```python
# understanding relationships between features(corelation)
wine.corr()
"""
+1 mean perfect positive corelation
-1 mean perfect negative corelation
any value between the -1 to +1 if value is much closer to 1 means stronger to the positive relation and closer to -1
strong relation to the negative
"""
# The .corr() method in Pandas computes the correlation matrix for all numerical columns in the DataFrame
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))

# sns.heatmap(wine.corr())
# seeing the relation of varaible in head map 
# we can also see the value here in heatmap
sns.heatmap(wine.corr(),annot=True)
```

```python
# let's see the quality column 
wine['quality'].value_counts()

# here all the data are not balance so we conclude that imbalance dataset
# quality
# 6    1788
# 5    1175
# 7     689
# 4     153
# 8     131
# 3      20
# 9       5
# Name: count, dtype: int64

# visualize with plotting it 
plt.xlabel("Wine Quality")
plt.ylabel("Values")
plt.title("Distirbution of wine quality show")
wine['quality'].value_counts().plot(kind='bar')
```
