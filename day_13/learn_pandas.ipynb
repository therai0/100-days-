```python
# Pandas is an open-source software library designed for data manipulation and analysis. It provides data structures like series and DataFrames to easily clean, transform and analyze large datasets and integrates with other Python libraries, such as NumPy and Matplotlib.
```

```python
import pandas as pd

"""
A pandas Series is a one-dimensional array-like object that can hold any data type. It is similar to a column in a table.
"""

# creating the Series
data = [ "Bhaskar","Sandesh","Manish","Bishal"]
ser = pd.Series(data)

ser
```

```python
# series of dictionary
dit = {
    "username":"Bhaskar",
    "password":"123434",
    "email":"bhaskar@gmail.com"
}
ser = pd.Series(dit)
print(f"Default index:\n{ser}")


items = ["apple","carrot","pork"]
# passing different value as index
ser = pd.Series(items,index=["fruits","vegetable","meat"])
print(f"Manual indexing:\n{ser}")
```

```python
""" Dataframes """

""" A DataFrame is a table-like data structure that consists of rows and columns, similar to a spreadsheet or SQL table."""


dit = {
    "richest_person":["Bhaskar Rai","Bishal Kafle","Ayush Dahal"],
    "networth":["10.5B","9.1B","2.5B"],
    "company":["BnB tech","Xyz Inc","Dahal Group"],
}

# creating the dataframe 

df = pd.DataFrame(dit)
df
```

```python
# Accessing data in dataframes

data = {
    "student":["Bhaskar","Aayush","Sandesh","Manish"],
    "class":[7,7,7,7],
    "age":[12,13,12,14],
    "address":["Mechinagar-3","Birtamode-4","Satashi-3","Kankai-1"]
    }

df = pd.DataFrame(data)

df 
```

```python
# accessing the element of dataframes
df["student"]
# access the all the data of name column
```

```python
# accesing all the data of 0th row
df.loc[0]

# Access a group of rows and columns by label(s) or a boolean array.
```

```python
# accessing the data of 0th row and 0th index
df.iloc[0,0]
# Purely integer-location based indexing for selection by position.
```

```python
# Access a single value for a row/column label pair.
df.at[1,"address"]
```

```python
# Access a single value for a row/column pair by integer position.
df.iat[0,0]
```

```python
# Data Manipulation with DataFrames

# adding new column in dataframe
df['father_name'] = ["Basu Dev","Hari","Kiran","Gaurav"] #it will add the new column of father_name

df
```

```python
# removing the column from the dataframes
df.drop("age",axis=1, inplace=True)
df
# this code will remove the age column from the dataframe
```

```python
# removing the row
df.drop(0,inplace=True)
df
# it will remove the first row from the dataframe
```
