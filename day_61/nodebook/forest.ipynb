```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df.head()
```

## Data Cleaning 

```python
# checking missing value 
df[df.isnull().any(axis=1)]
```

This dataset belong to the two region Bejaia region and Sidi-Bel Abbes region so we need to seperate the this two region. So Bejaia region will be 0 and Sidi-Bel Abbes region will be 1. 


```python
df.loc[:122,'Region']= 0
df.loc[122:,'Region']= 1
```

```python
df.tail()
```

```python
# converting their datatypes into integer
df['Region'] = df['Region'].astype(int)
```

```python
# let's drop the all nan value because there is few nan valus in dataset
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
# I think there is something in 122 row 
df.iloc[[122]]
```

```python
# in the 122 row there column header so we need to drop this also
df = df.drop([122]).reset_index(drop=True)
```

```python
# check the column names 
df.columns
```

```python
# so let's remove the all the extra spaces from the column names
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

```python
# let's convert the datatype of different column into integer and float
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]].astype(int)
df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI',]] = df[['Rain', 'FFMC','DMC', 'DC', 'ISI', 'BUI', 'FWI',]].astype(float)
```

```python
df.info()
```

```python
df.to_csv('clean_dataset.csv')
```
