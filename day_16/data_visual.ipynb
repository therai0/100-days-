```python
import seaborn as sns

sns.__version__

""" 
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
"""
```

```python
tips_data = sns.load_dataset('tips')
tips_data
```

```python
# ploting the data total bill per day

per_dat_total_bill = tips_data.groupby('day')['total_bill'].sum()
per_dat_total_bill
```

```python
""" Line Plot """

import matplotlib.pyplot as plt 

days = list(set(tips_data['day']))
sns.lineplot(y=per_dat_total_bill,x=days)
plt.title("Line plot")
plt.xlabel("Days")
plt.ylabel("Total bill")
plt.show()
```

```python
penguins_data_set = sns.load_dataset('Penguins')
penguins_data_set
```

```python
""" Scatter plot """

# x-axis will be the length of the bill
# y-axis will be the body mass of the of penguins

sns.scatterplot(x='bill_length_mm',y='body_mass_g',data=penguins_data_set)
plt.title('Mass of penguins according to their bill')
plt.xlabel('Length of bill in MM')
plt.ylabel('Mass of body in kilogram')
plt.grid()
plt.show()
```

```python
# bar plot 
titanic_data = sns.load_dataset('titanic')
titanic_data
```

```python
"""
Categorical plot
"""
sns.barplot(x='class',y='age',data=titanic_data)
plt.xlabel('Seat class')
plt.ylabel('Age of passenger')
plt.title('Titanic data')
plt.show()
```

```python
"""
Box Plot
Box plots are useful for identifying outliers and understanding the distribution of data.
"""
flights_dataset = sns.load_dataset('flights')
flights_dataset
```

```python
# box ploting of year and passengers

sns.boxplot(x='year',y='passengers',data=flights_dataset)
plt.xlabel('Total number of passengers')
plt.ylabel('Years')
plt.title('Box ploting of number of passenger per year')
plt.show()
```

```python
""" 
Histograms and KDE Plots
KDE ==>  Kernel Density Estimation
Kernel Density Estimate (KDE) Plot is a powerful tool for estimating the probability density function of continuous or non-parametric data. KDE plot is implemented through the kdeplot function in Seaborn.
"""
# histograms of penguins
sns.histplot(data=penguins_data_set,x='body_mass_g')
plt.xlabel('Penguin body mass in gram')
plt.ylabel('Number of penguins')
plt.title('Histogras of penguins')
plt.show()
```

```python
# KDE
sns.kdeplot(penguins_data_set['flipper_length_mm'],fill=True)
```

```python
"""
Pair Plot
Pair plots are useful for visualizing relationships between all numerical variables in the dataset.
"""

# it show the corelation of all numerical variable in dataset
sns.pairplot(penguins_data_set)
```

```python
# Correlation and Heatmap
import seaborn as sns
penguins_data_set = sns.load_dataset('penguins')
numeric_values = ['bill_length_mm','bill_depth_mm','body_mass_g']
co_relations = penguins_data_set[numeric_values].corr()

co_relations
```

```python
# creating the heatmap
sns.heatmap(co_relations,annot=True)
```
