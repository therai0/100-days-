```python
import pandas as pd

from io import StringIO

print(pd.__version__)
data_d = '''
[{
    "albumId": 1,
    "id": 1,
    "title": "accusamus beatae ad facilis cum similique qui sunt",
    "url": "https://via.placeholder.com/600/92c952",
    "thumbnailUrl": "https://via.placeholder.com/150/92c952"
  },
  {
    "albumId": 1,
    "id": 2,
    "title": "reprehenderit est deserunt velit ipsam",
    "url": "https://via.placeholder.com/600/771796",
    "thumbnailUrl": "https://via.placeholder.com/150/771796"
  }]
'''

str_obj = StringIO(data_d) # it convert the data into string object

df = pd.read_json(str_obj)
df
```

```python
# converting back to json data
df.to_json()
```

```python
# converting back to json with orient is index
df.to_json(orient='index')
```

```python
# converting back to json with orient is record
df.to_json(orient='records')
# it will convert in orginal form
```
