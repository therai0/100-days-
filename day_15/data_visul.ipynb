```python
""" Data Visualization With Matplotlib """


import matplotlib.pyplot as plt 

x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]

plt.plot(x,y)
plt.title("Line plit")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]

plt.title("Line plit")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

# we can customize the plot line 
plt.plot(x,y,color='black',linestyle='', marker = 'o')
plt.grid() #it helps to show the gird also
plt.show()
```

```python
# we can create the multiple plot at single time

x = [1,2,3,4,5,6]
y = [1,4,9,16,25,36]

x2 = [1,3,5,7,11]
y2 = [1,9,25,49,121]

plt.figure(figsize=(16,4)) # plotting canves
# Creates a new figure window (i.e., a blank plot canvas) and set the height and width

# for plot one
plt.subplot(1,2,1) # this means position of plot where it should be created first argument is number of rows second argument is column (how many column) and third argument is position where you need to plot this figure. there will be one row and two column and figure will be made in first column

plt.plot(x,y,color='red',linestyle='',marker='o')
plt.title('Plot 1')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# for plot two
plt.subplot(1,2,2)
plt.plot(x2,y2,color='black',linestyle='',marker='x')
plt.title('Plot 1')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")


```

```python
# Bar plot 

users = [1.5,7,3,5,6]
apps = ['facebook','tiktok','instagram','snapchat','x']

plt.bar(apps,users)
plt.xlabel('social media')
plt.ylabel('Users in billion')
```

```python
# Histogram
# Histograms are used to represent the distribution of numerical data.

population = [10,15,20,25,30,35,45]

plt.hist(population,bins=5,color='skyblue',edgecolor='white')
plt.grid()
plt.show()
```

```python
# pie chart
owners = ['Bhaskar', 'Aayush', 'Bishal', 'Manish']
shares = [30,20,30,20]
colors = ['lightblue','skyblue','pink','red']

plt.pie(shares,labels=owners,colors=colors,explode=(0.1,0,0,0),autopct='%1.1f%%')
# exploded  means moving that particular slice 
```
