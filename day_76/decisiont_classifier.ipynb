## Decision Tree Classifier

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris_dataset
```

```python
iris_dataset.target
```

```python
print(iris_dataset['DESCR'])
```

```python
# independent feature
x = pd.DataFrame(iris_dataset.data, columns=['sepal length','sepal width','petal length','petal widt'])
x.head()
```

```python
y = iris_dataset.target
```

```python
# spliting in train and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
# traning model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(x_train,y_train)
y_pred = dt_classifier.predict(x_test)
metrix = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
print(metrix)
print(report)
```

```python
# visiualing in binary tree 
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(dt_classifier,filled=True)
```

## Post Prepruning


```python
# To overcome the overfitting change the number of depth
# traning model
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,confusion_matrix
dt_classifier = DecisionTreeClassifier(max_depth=2)
dt_classifier.fit(x_train,y_train)
y_pred = dt_classifier.predict(x_test)
metrix = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
print(metrix)
print(report)
```

```python
# visiualing in binary tree 
from sklearn import tree
plt.figure(figsize=(10,7))
tree.plot_tree(dt_classifier,filled=True)
```

## Decision Tree Classifier Pre-pruning and Hyperparameter tuning

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
dt_model = DecisionTreeClassifier()
grid_cv = GridSearchCV(estimator=dt_model,param_grid=params,refit=True,cv=5)
grid_cv.fit(x_train,y_train)
y_pred = grid_cv.predict(x_test)
metrix = confusion_matrix(y_test,y_pred)
report = classification_report(y_test,y_pred)
print(metrix)
print(report)
```
