```python
import seaborn as sns 
import pandas as pd 

data = sns.load_dataset('dots')

df = pd.DataFrame(data)
```

```python
df.isnull()
# there is no null value
```

```python
# finding the mean,median and mode 
mean = df['coherence'].mean()
print(mean)
median = df['coherence'].median()
print(median)
mode = df['coherence'].mode()
print(mode)
```

```python
zero_nums  = df[df['coherence'] == 0]
print(len(zero_nums))
print(df.shape)
```

```python
import matplotlib as plt 
penguin = sns.load_dataset('penguins')

df_2 = pd.DataFrame(penguin)
df_2
```

```python
df_2['body_mass_g'].isnull().value_counts()
# output = 2 
# there is 2 row have NaN value so we are going to replace that NaN with mean value
df_2['updated_body_mass']=df_2['body_mass_g'].fillna(df_2['body_mass_g'].mean())

```

```python
df_2['updated_body_mass'].isnull().value_counts()
# new we create have new colum with null value is 0
```
