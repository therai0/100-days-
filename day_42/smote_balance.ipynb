```python
from  sklearn.datasets import make_classification

# Create an imbalanced dataset with 5000 samples, 2 features, 1 cluster per class, and class weights to induce imbalance
x, y = make_classification(n_samples=5000, n_features=2, n_clusters_per_class=1, weights=[0.60], random_state=12, n_redundant=0)
```

```python
import pandas as pd

# Create DataFrames for features and target
df1 = pd.DataFrame(x, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])

# Concatenate features and target into a single DataFrame
final_df = pd.concat([df1, df2], axis=1)
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt

# let's visualize the data
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Imbalanced Dataset Visualization')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE oversampler
oversample = SMOTE()


x_resampled, y_resampled = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```
